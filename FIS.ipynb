```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

# Define input variables
motivation = ctrl.Antecedent(np.arange(0, 101, 1), 'motivation')
possession = ctrl.Antecedent(np.arange(0, 101, 1), 'possession')
equipment = ctrl.Antecedent(np.arange(0, 101, 1), 'equipment')

# Define output variable
coal_production = ctrl.Consequent(np.arange(0, 101, 1), 'coal_production')

# Define linguistic terms and membership functions for each variable
motivation['low'] = fuzz.trimf(motivation.universe, [0, 0, 40])
motivation['medium'] = fuzz.trimf(motivation.universe, [40, 50, 60])
motivation['high'] = fuzz.trimf(motivation.universe, [60, 100, 100])

possession['not_available'] = fuzz.trimf(possession.universe, [0, 0, 80])
possession['partially_available'] = fuzz.trimf(possession.universe, [80, 90, 90])
possession['highly_available'] = fuzz.trimf(possession.universe, [90, 100, 100])

equipment['low'] = fuzz.trimf(equipment.universe, [0, 0, 70])
equipment['medium'] = fuzz.trimf(equipment.universe, [70, 78, 78])
equipment['high'] = fuzz.trimf(equipment.universe, [80, 100, 100])

coal_production['very_low'] = fuzz.trimf(coal_production.universe, [0, 0, 20])
coal_production['low'] = fuzz.trimf(coal_production.universe, [20, 40, 60])
coal_production['medium'] = fuzz.trimf(coal_production.universe, [40, 60, 80])
coal_production['high'] = fuzz.trimf(coal_production.universe, [60, 80, 100])
coal_production['very_high'] = fuzz.trimf(coal_production.universe, [80, 100, 100])

# Define the fuzzy rules
rules = [
    ctrl.Rule(motivation['low'] & possession['not_available'] & equipment['low'], coal_production['very_low']),
    ctrl.Rule(motivation['low'] & possession['not_available'] & equipment['medium'], coal_production['low']),
    ctrl.Rule(motivation['low'] & possession['not_available'] & equipment['high'], coal_production['medium']),

    ctrl.Rule(motivation['low'] & possession['partially_available'] & equipment['low'], coal_production['low']),
    ctrl.Rule(motivation['low'] & possession['partially_available'] & equipment['medium'], coal_production['medium']),
    ctrl.Rule(motivation['low'] & possession['partially_available'] & equipment['high'], coal_production['high']),

    ctrl.Rule(motivation['low'] & possession['highly_available'] & equipment['low'], coal_production['medium']),
    ctrl.Rule(motivation['low'] & possession['highly_available'] & equipment['medium'], coal_production['high']),
    ctrl.Rule(motivation['low'] & possession['highly_available'] & equipment['high'], coal_production['high']),

    ctrl.Rule(motivation['medium'] & possession['not_available'] & equipment['low'], coal_production['low']),
    ctrl.Rule(motivation['medium'] & possession['not_available'] & equipment['medium'], coal_production['medium']),
    ctrl.Rule(motivation['medium'] & possession['not_available'] & equipment['high'], coal_production['high']),

    ctrl.Rule(motivation['medium'] & possession['partially_available'] & equipment['low'], coal_production['medium']),
    ctrl.Rule(motivation['medium'] & possession['partially_available'] & equipment['medium'], coal_production['high']),
    ctrl.Rule(motivation['medium'] & possession['partially_available'] & equipment['high'], coal_production['high']),

    ctrl.Rule(motivation['medium'] & possession['highly_available'] & equipment['low'], coal_production['high']),
    ctrl.Rule(motivation['medium'] & possession['highly_available'] & equipment['medium'], coal_production['high']),
    ctrl.Rule(motivation['medium'] & possession['highly_available'] & equipment['high'], coal_production['high']),

    ctrl.Rule(motivation['high'] & possession['not_available'] & equipment['low'], coal_production['medium']),
    ctrl.Rule(motivation['high'] & possession['not_available'] & equipment['medium'], coal_production['high']),
    ctrl.Rule(motivation['high'] & possession['not_available'] & equipment['high'], coal_production['high']),

    ctrl.Rule(motivation['high'] & possession['partially_available'] & equipment['low'], coal_production['high']),
    ctrl.Rule(motivation['high'] & possession['partially_available'] & equipment['medium'], coal_production['high']),
    ctrl.Rule(motivation['high'] & possession['partially_available'] & equipment['high'], coal_production['high']),

    ctrl.Rule(motivation['high'] & possession['highly_available'] & equipment['low'], coal_production['high']),
    ctrl.Rule(motivation['high'] & possession['highly_available'] & equipment['medium'], coal_production['high']),
    ctrl.Rule(motivation['high'] & possession['highly_available'] & equipment['high'], coal_production['very_high']),
]

# Create the control system
coal_production_ctrl = ctrl.ControlSystem(rules)
coal_production_sim = ctrl.ControlSystemSimulation(coal_production_ctrl)

# Set input values (adjust these values as needed)
coal_production_sim.input['motivation'] = 70
coal_production_sim.input['possession'] = 85
coal_production_sim.input['equipment'] = 75

# Compute the result
coal_production_sim.compute()

# Print the result
print("Coal Production:", coal_production_sim.output['coal_production'])

# Visualize the membership functions and rules
motivation.view(sim=coal_production_sim)
possession.view(sim=coal_production_sim)
equipment.view(sim=coal_production_sim)
coal_production.view(sim=coal_production_sim)

# Show the plots
plt.show()
```
