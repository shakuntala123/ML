```python
X=data[:,0:10]
X
```

```python
X.shape
X
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,1))
```

```python
#fit_transform() is used to both fit the scaler to the data and transform the data to the scaled values.
x =min_max_scaler.fit_transform(X)
x
```

```python
sd = preprocessing.StandardScaler()
x_after_sd = sd.fit_transform(X)
x_after_sd
```

```python
#calculating mean
X_mean = X.mean()
#claculating standard deviation
X_std = X.std()
#calculating Standardization
Z = (X -X_mean)/X_std
```

```python
from sklearn.decomposition import PCA

# Create principal components
pca = PCA()
X_pca = pca.fit_transform(Z)

# Convert to dataframe
component_names = [f"PC{i+1}" for i in range(X_pca.shape[1])]
X_pca = pd.DataFrame(X_pca, columns=component_names)

X_pca.head()
```

```python
import pandas as pd
import numpy as np

# Create a DataFrame of loadings
loadings = pd.DataFrame(
    pca.components_.T,        # Transpose of the matrix of loadings
    columns=component_names,  # Names of the principal components
    index=np.arange(X.shape[1])  # Indices of the original features
)
loadings
```
